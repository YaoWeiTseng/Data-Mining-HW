```python
import pandas as pd
data = pd.read_csv('HW2DATA.csv')

#取出10~12月資料
data = data[4908:6564]
data = data.reset_index()
data = data.drop(labels=["index"], axis=1)

#取代NR為0
data.replace('NR',0, inplace=True)
data
```

```python
# 缺失值以及無效值以前後一小時平均值取代
for i in range(len(data)):
    for j in range(3,27):
            x = i
            y = j
            if str.find(str(data.iloc[i,j]),'#')!= -1 or str.find(str(data.iloc[i,j]),'*')!= -1 or str.find(str(data.iloc[i,j]),'x')!= -1 or str.find(str(data.iloc[i,j]),'A')!= -1 or pd.isnull(data.iloc[i,j]) == True:
                if j != 3 and j != 26:
                    if str.find(str(data.iloc[i,j+1]),'#')== -1 and str.find(str(data.iloc[i,j+1]),'*')== -1 and str.find(str(data.iloc[i,j+1]),'x')== -1 and str.find(str(data.iloc[i,j+1]),'A')== -1 and pd.isnull(data.iloc[i,j+1])== False: #不包含
                        data.iloc[i,j] = (float(data.iloc[i,j-1])+float(data.iloc[i,j+1]))/2
                    else:
                        while str.find(str(data.iloc[x,y]),'#')!= -1 or str.find(str(data.iloc[x,y]),'*')!= -1 or str.find(str(data.iloc[x,y]),'x')!= -1 or str.find(str(data.iloc[x,y]),'A')!= -1 or pd.isnull(data.iloc[x,y]) == True:
                            y += 1
                            if y == 27:
                                x += 18
                                y = 3
                        data.iloc[i,j] = (float(data.iloc[i,j-1])+float(data.iloc[x,y]))/2
                elif j == 3:
                    if str.find(str(data.iloc[i,j+1]),'#')== -1 and str.find(str(data.iloc[i,j+1]),'*')== -1 and str.find(str(data.iloc[i,j+1]),'x')== -1 and str.find(str(data.iloc[i,j+1]),'A')== -1 and pd.isnull(data.iloc[i,j+1]) ==False: 
                        data.iloc[i,j] = (float(data.iloc[i-18,26])+float(data.iloc[i,j+1]))/2
                    else:
                        while str.find(str(data.iloc[x,y]),'#')!=-1 or str.find(str(data.iloc[x,y]),'*')!=-1 or str.find(str(data.iloc[x,y]),'x')!=-1 or str.find(str(data.iloc[x,y]),'A')!=-1 or pd.isnull(data.iloc[x,y])==True:
                            y+=1
                            if y==27:
                                x+=18
                                y = 3
                        data.iloc[i,j] = (float(data.iloc[i-18,26])+float(data.iloc[x,y]))/2
                elif j == 26:
                    while str.find(str(data.iloc[x+18,29-y]),'#')!= -1 or str.find(str(data.iloc[x+18,29-y]),'*')!=-1 or str.find(str(data.iloc[x+18,29-y]),'x')!=-1 or str.find(str(data.iloc[x+18,29-y]),'A')!=-1 or pd.isnull(data.iloc[x+18,29-y])==True:
                        y -= 1
                        if y == 2:
                            x += 18
                            y = 26
                    data.iloc[i,j] = (float(data.iloc[i,j-1])+float(data.iloc[x+18,29-y]))/2
            else:
                continue
```

```python
data = data.drop(['日期', '測站', '測項'], axis=1)
```

```python
#分割訓練集並製成時序資料
train = data[0:18]
for i in range(18, 1098, 18):
    split = data.iloc[i:i+18,:]
    split.index = range(0, 18)
    train = pd.concat([train, split], axis = 1, join_axes = [train.index])
ind = []
for i in range(len(train.columns)):
    ind.append(str(i))
train.columns = ind    
train
```

```python
#分割測試集並製成時序資料
test = data.iloc[1098:1116,:]
for i in range(1116, 1656, 18):
    split = data.iloc[i:i+18,:]
    split.index = range(1098, 1116)
    test = pd.concat([test, split], axis = 1, join_axes = [test.index])
ind = []
for i in range(len(test.columns)):
    ind.append(str(i))
test.columns = ind    
test
```

```python
import numpy as np
trainX = np.array(train)[9] #訓練集中PM2.5的數值
X_train = []
for i in range(0, len(trainX)-6,1):
    X_train.append(trainX[i:i+6])

y_train =[]
for i in range(6, len(trainX)):
    y_train.append(trainX[i])
    
testX = np.array(test)[9] #測試集中PM2.5的數值
X_test = []
for i in range(0, len(testX)-6,1):
    X_test.append(testX[i:i+6])

y_test =[]
for i in range(6, len(testX)):
    y_test.append(float(testX[i]))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
model1 = LinearRegression()
model1.fit(X_train, y_train)
pred1 = model1.predict(X_test)
print("以PM2.5預測PM2.5之線性回歸MAE分數:",mean_absolute_error(y_test, pred1))

model2 = RandomForestRegressor(500)
model2.fit(X_train, y_train)
pred2 = model2.predict(X_test)
print("以PM2.5預測PM2.5隨機森林回歸MAE分數:",mean_absolute_error(y_test, pred2))
```
